```python
import numpy as np
import cv2
import os
import tensorflow as tf
```

```python
from keras.preprocessing.image import ImageDataGenerator
tf.keras.preprocessing.image.img_to_array

from PIL import Image
```

```python
X =  []

for img_name in os.listdir("./gdrive/MyDrive/GCSS/little_finger"):
  image = Image.open(f'./gdrive/MyDrive/GCSS/little_finger/{img_name}')
  rgb_img=image.convert('RGB')
  X.append(rgb_img)
```

```python
len(X)
```

```python
for img in X:
    # Initialising the ImageDataGenerator class.
    # We will pass in the augmentation parameters in the constructor.
    datagen = ImageDataGenerator(
            #rotation_range = 40,
            shear_range = 0.2,
            zoom_range = 0.2,
            #horizontal_flip = True,
            )

    # Loading a sample image
    #img = load_img('image.jpg')
    # Converting the input sample image to an array
    x = tf.keras.preprocessing.image.img_to_array(img, data_format=None, dtype=None)
    # Reshaping the input image
    x = x.reshape((1, ) + x.shape)

    # Generating and saving 5 augmented samples
    # using the above defined parameters.
    i = 0
    for batch in datagen.flow(x, batch_size = 1,
                              save_to_dir ='gdrive/MyDrive/aug_avnish/little',
                              save_prefix ='image', save_format ='jpg'):
        i += 1
        if i > 1:
            break
```

```python
from numpy import expand_dims
tf.keras.preprocessing.image.img_to_array
# import cv2
# from PIL import Image
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot as plt
import matplotlib.image as mpimg


# we first load the image
from PIL import Image
```

```python
for pic in X:
  dataImage = tf.keras.preprocessing.image.img_to_array(pic)
  # print(dataImage)
  # expanding dimension of the load image
  imageNew = expand_dims(dataImage, 0)
  # now here below we creating the object of the data augmentation class
  imageDataGen = ImageDataGenerator(brightness_range=[0.4,0.6])
  # because as we alreay load image into the memory, so we are using flow() function, to apply transformation
  iterator = imageDataGen.flow(imageNew, batch_size=1)


  # below we generate augmented images and plotting for visualization


  for i in range(3):
    # we are below define the subplot
    ##plt.subplot(330 + 1 + i)
    # generating images of each batch
    batch = iterator.next()
    # again we convert back to the unsigned integers value of the image for viewing
    image = batch[0].astype('uint8')

    #print(type(image))
    # creating image object of
    # above array
    data = Image.fromarray(image)
    data.save(f"./gdrive/MyDrive/aug_avnish/little/{pic}{i}.jpg")
   # data.save(f"./gdrive/MyDrive/data_open_brt/{pic}{i}.jpeg")
```

```python
import os
import shutil

source_dir = './gdrive/MyDrive/train/Open'
dest_dir = './gdrive/MyDrive/new_aug/open'

# Create the destination directory if it doesn't exist
if not os.path.exists(dest_dir):
    os.makedirs(dest_dir)

# Loop through all files in the source directory
for filename in os.listdir(source_dir):
    # Check if the file is an image (JPEG, PNG, or JPG)
    if filename.endswith('.jpg') or filename.endswith('.png') or filename.endswith('.jpeg'):
        # Construct the full file path
        filepath = os.path.join(source_dir, filename)
        # Copy the file to the destination directory
        shutil.copy(filepath, dest_dir)
```
